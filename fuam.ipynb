#### Funciones de distancia


```python
import math
import pandas as np

#
# Euclidean distance
def euclidean_distance(point1, point2):
    sum_squared_diff = sum((p1 - p2) ** 2 for p1, p2 in zip(point1, point2))
    return math.sqrt(sum_squared_diff)

# Manhattan distance
def manhattan_distance(point1, point2):
    return sum(abs(p1 - p2) for p1, p2 in zip(point1, point2))

# Cosine similarity
def cosine_similarity(point1, point2):
    dot_product = sum(p1 * p2 for p1, p2 in zip(point1, point2))
    magnitude1 = math.sqrt(sum(p ** 2 for p in point1))
    magnitude2 = math.sqrt(sum(p ** 2 for p in point2))
    if magnitude1 == 0 or magnitude2 == 0:
        return 0  
    return dot_product / (magnitude1 * magnitude2)
```

##Datos

```python
import pandas as pd

DataInicial = pd.read_csv("./files/Vinos.csv")
atributosName = DataInicial.columns[:-1]
atributosClase = DataInicial.columns[-1]

Datos = DataInicial[atributosName]
Clase = DataInicial[atributosClase]

print("Column names:", DataInicial.columns)
print(Clase.value_counts())
```

```python
from sklearn.model_selection import train_test_split

#Obtain my test and train values
datosTrain, datosTest, claseTrain, claseTest =  train_test_split(Datos, Clase, test_size=0.30)

print("CONJUNTOS DE ENTRENAMIENTO Y TEST\n")
print("Instancias de conjunto de entrenamiento")
print(claseTrain.value_counts(), "\n")

print("Instancias de conjunto para clasificar")
print(claseTest.value_counts(), "\n")
```

Encuentra vecinos

```python
# Step 4: Calculate Distances
def get_distances(test_instances, training_data):
    distances = []
    for _, test_instance in test_instances.iterrows():
        for _, train_instance in training_data.iterrows():
            dist = euclidean_distance(test_instance[:-1], train_instance[:-1])
            distances.append((train_instance, dist))
    return distances



# Step 5: Find Nearest Neighbors
def get_neighbors(test_data, train_data, k):
    neighbors_list = []
    distances = get_distances(test_data, train_data)
    for i in range(len(test_data)):
        test_instance = test_data.iloc[i]
        test_instance_distances = distances[i*len(train_data):(i+1)*len(train_data)]
        test_instance_distances.sort(key=lambda x: x[1])
        neighbors_list.append(test_instance_distances[:k])
    return neighbors_list


# Step 6: Make Predictions
def predict_classification(neighbors_list):
    predictions = []
    for neighbors in neighbors_list:
        votes = {}
        for neighbor in neighbors:
            print("Neighbor:", neighbor)  # Print neighbor to see its content
            label = neighbor[0][atributosClase]# Access the target variable from the training instance
            if label in votes:
                votes[label] += 1
            else:
                votes[label] = 1
        predictions.append(max(votes, key=votes.get))
    return predictions




# Step 5: Find Nearest Neighbors
k = 3
neighbors_list = get_neighbors(datosTest, datosTrain, k)

# Step 6: Make Predictions
predictions = predict_classification(neighbors_list)
print("Predicted classes:", predictions)
```
