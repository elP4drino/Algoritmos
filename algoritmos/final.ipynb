### 🌳 Información Random Forest Classifier

##### ⚙️ ¿Cómo funciona?
Este algoritmo de clasificación genera un conjunto de árboles de decisión independientes que toman diversos muestreos de los datos, asi como atributos o columnas del dataset original, generando de este modo una serie de múltipes escenarios que se pueden evaluar de manera conjunta al resto de árboles dentro del bosque, generando una predicción más exacta en base a todos los escenarios.

A la generación de múltiples árboles de decisión con atributos random y diferentes se le conoce oficialmente como bagging, mientras que una vez que ya se cuentan con múltiples árboles acumulados se realiza el voting, se trata de un proceso de conteo simple, en el que se calcula la cantidad de escenarios o árboles que predicen una clase u otra, llegando asi a dar como predicción general del bosque, a la predicción que más aparezca dentro de todos los árboles.

Modelo de Clasificación.

#### ⚙️ Tipos de datos que acepta
Acepta datos categoricos como objetivo, mientras que numericos en el contenido de cada uno de los atributos del dataset, es decir, es capaz de recibir la misma clase de valores que otros modelos de clasificación como el Knn o árboles de decisión simples.

#### ⚙️ Parámetros
- "n_estimadores": El número de árboles en el ensemble. 
- "max_depth": Indicar la profundidad de los árboles dentro dle bosque
- "min_samples_split": Indica el número mínimo de muestras para ser dividido
- "random_state": Genera  un valor aleatorio para iniciar con el generador de números pseudoaleatorios y poder reproducir escenarios a pesar de ser randoms.
